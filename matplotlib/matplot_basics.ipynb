```python
x = np.arange(1,101)
fig,axs = plt.subplots(2,2,figsize=(12,6))
axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title('sin(x)')

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title('cos(x)')

axs[1,0].plot(x,np.random.random(100))
axs[1,0].set_title('random')

axs[1,1].plot(x,np.log(x))
axs[1,1].set_title('log(x)',fontsize=20,fontname="algerian")
fig.suptitle('four plot', fontsize=20,fontname="algerian",color="cyan")
plt.subplots_adjust(wspace=0.5, hspace=0.5)  # Adjust width and height spacing
plt.savefig('four_plot.png',dpi=300)
```

```python
ax = plt.axes(projection='3d')
p = np.arange(-5,5,0.1)
q = np.arange(-5,5,0.1)

P,Q = np.meshgrid(p,q)

R=np.sin(P)*np.cos(Q)
ax.plot_surface(P,Q,R,cmap='Spectral')
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Importing 3D toolkit

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Define 3D axes

p = np.arange(-5, 5, 0.1)
q = np.arange(-5, 5, 0.1)

P, Q = np.meshgrid(p, q)

R = np.sin(P) * np.cos(Q)  # Corrected formula

ax.plot_surface(P, Q, R, cmap='viridis')  # Added colormap for better visualization

plt.show()
```

```python
import random
heads_tails =[0,0]

for _ in range(10000):
    heads_tails[random.randint(0,1)] += 1
    plt.bar(['heads','tails'],heads_tails,color=['red','blue'])
    plt.pause(0.001)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Initialize the figure and bar chart
fig, ax = plt.subplots()
counts = [0, 0]
bars = ax.bar(['heads', 'tails'], counts, color=['red', 'blue'])
ax.set_ylim(0, 1000)  # Set a limit to avoid rescaling

# Loop to simulate coin flips
for _ in range(2000):
    counts[random.randint(0, 1)] += 1
    for i, bar in enumerate(bars):
        bar.set_height(counts[i])
    plt.pause(0.001)  # Pause briefly to update the plot

plt.show()
```
